## 1. Introduction

Dask is a parallel and distributed computing Python libary. In a nutshell, Dask allows you to:

- set up a **task graph** from several APIs (arrays, data frames, tasks, etc.).
- **run the graph in parallel** on different types of infrastructure (multi-core CPUs, cloud, HPC).

## 2. Creating a task graphs

### 2.1 Dask Delayed

The [Dask delayed](https://docs.dask.org/en/stable/delayed.html) interface is generic, and it allows to build task graphs directly from Python functions.

```python
from dask.delayed import delayed
```

```python
@delayed
def add(x, y):
    print(x, " + ", y)
    return x + y
```

```python
a_p = add(1, 2)
```

```python
b_p = add(a_p, 3)
```

```python
c_p = add(a_p, b_p)
```

Visualize task graph using `dask.visualize()`:

```python
import dask
dask.visualize(c_p, rankdir="LR")
```

Execute the graph (using the default thread-based scheduler) with the `.compute()` method:

```python
c_p.compute()
```

### 2.2 Dask Arrays

When working with more complex data structures, we can use Dask high-level APIs - e.g. for [array data](https://docs.dask.org/en/stable/array.html). 

From the [Dask documentation](https://docs.dask.org/en/stable/array.html):

<img src="https://docs.dask.org/en/stable/_images/dask-array.svg" width=70%>



```python
import dask.array as da
x = da.random.random((2000, 1000), chunks=(500, 500))
x
```

```python
y = da.dot(x, x.T)
```

```python
z = y.mean()
```

Again, calling `dask.visualize()` on a Dask object shows the task graph:

```python
dask.visualize(z)
```

And again, calling the `.compute()` method execute the task graph (with the default thread-based scheduler).

```python
z.compute()
```

### 2.3 Xarray

[Xarray](https://docs.xarray.dev) is the most popular Python library for (labelled) multi-dimensional arrays. 

From the [Xarray documentation](https://docs.xarray.dev): 

<img src="https://docs.xarray.dev/en/stable/_images/dataset-diagram.png" width=70%>

Xarray supports Dask Array as internal data structure, enabling parallel computing and handling of larger-than-memory datasets. Xarray has two main data structures:

* `xarray.DataArray`, an array with labelled dimensions and coordinates. 
* `xarray.Dataset`, a collection of DataArray's with shared dimensions and coordinates.

```python
import xarray as xr

nc_path = '/project/remotesensing/Share/hpcdat_ws1/daymet-daily-v4/region-hi/daymet_v4_daily_hi_tmax_1980.nc'

dataset = xr.open_dataset(nc_path, engine="h5netcdf", chunks={})
dataset
```

```python
tmax = dataset["tmax"]
tmax
```

As for Dask Array objects, calculations are executed lazily, and only the data that is needed is actually loaded into memory:

```python
subset = tmax.sel(time="1980-01-01")
subset_max = subset.max()
```

Similarly to Dask Array objects, also Xarray objects have the `.compute()` method:

```python
subset_max.compute()
```

Some methods automatically trigger computation:

```python
subset.squeeze().plot.imshow()
```

## 3. Execute task graphs

Dask supports [several types of infrastructure](https://docs.dask.org/en/stable/deploying.html) to execute the task graph.

### 3.1 Multi-threading/processing

One can run the graph locally using `multithreading` or `multiprocessing`:

```python
subset_max.compute(scheduler="threads", n_workers=2)
```

```python
subset_max.compute(scheduler="processes", n_workers=2)
```

### 3.2 Distributed scheduler

## 4. Final note

Dask is not magic! It lowers the barrier to run big/parallel calculations, but the extend by which these will be efficient depend on many factors, e.g.:

* the **algorithm** you are trying to parallelize;
* the overall **size** of your dataset;
* the **chunking** strategy - both size and shape matter!

General advices:

* Start "serial" and try to scale up only if needed;
* Run benchmarks and/or profiling on your code (at the very least, check the Dask dashboard);
* "Blind scaling" (e.g. replacing `np` -> `da`) is typically not a good idea ...
